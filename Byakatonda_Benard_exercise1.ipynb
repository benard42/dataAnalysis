### 10. Create a new column called `financestatus` using the `map` method, such that for rows where the applicant income is less than 5,000 the value is `poor`, where the amount is between 5,000 and 20,000 the value is `middle class` and for anything above that, the value is `rich`

```python

def income_bins(ApplicantIncome):
    if ApplicantIncome in range(0,5000):
        return "poor"
    if ApplicantIncome in range(5000,20000):
        return "middle class"
    #if ApplicantIncome in range(20000,):
       # return "rich"
    else:
        return "rich"
```

```python
loans["financestatus"] = loans["ApplicantIncome"].map(income_bins)
loans.head(2)
```

### 11. Create a new column called `coapplicantstatus` such that for rows where the applicant income is less than 5,000 the value is `poor`, when the amount is between 5,000 and 20,000 the value is `middle class` and for anything above that, the value is `rich`. Do this by binning the data using the `pd.cut` method

```python
income_bins = [0,5000,20000,41667]
income_names = ["poor", "middle", "rich"]
loans["coapplicantstatus"] = pd.cut( loans["ApplicantIncome"], bins=income_bins, labels = income_names)
loans.head(10)
```

### 12. Create a new column called `previouscustomer` using the `Credit_History` column such that if `Credit_History` is `0` this column should contain `no`, if it is `1` then it should contain `yes`. If it is `null` or `NaN` it should return `NaN` (also call `null`) as well. 

*HINT:* In numpy, you can get the null value using `np.nan` or in Python using the keyword `None`

```python

def credit_bins(Credit_History):
    if Credit_History == 0:
        return "no"
    if Credit_History == 1:
        return "yes"
    #if Credit_History == NaN:
        #return "NaN"
    else:
        return "NaN"
```

```python
loans["previouscustomer"] = loans["Credit_History"].map(credit_bins)
loans.head(2)
```
